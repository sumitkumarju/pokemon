```python
from os import listdir
import pandas as pd
import PIL
from PIL import Image
from matplotlib import image
from matplotlib import pyplot
from numpy import asarray
import numpy as np
```

```python
loaded_images = list()
for filename in listdir('Train/Images'):
    image = Image.open('Train/Images/' + filename)
    gs_image = image.convert(mode='L')
    img_resized = gs_image.resize((100,100))
    data = asarray(img_resized)
    data = data.reshape((10000,1))
    loaded_images.append(data)
```

```python
df=pd.read_csv("Train/train.csv")
```

```python
data=df.values
```

```python
Y=data[:,1]
X=data[:,2:]
```

```python
len(X[0])
```

```python
def dist(a1,a2):
    return np.sum((a1-a2)**2)**.5
```

```python
def KNN(X,Y,test_point,k=3):
    m=X.shape[0]
    vals=[]
    for i in range(m):
        d=dist(test_point,X[i])
        vals.append((d,Y[i]))
    vals=sorted(vals)
    vals=np.array(vals[:k])
    b=np.unique(vals[:,1],return_counts=True)
    idx=np.argmax(b[1])
    pred=b[0][idx]
    return pred
```

```python
from collections import namedtuple

Pair = namedtuple("Pair", ["first", "second"])

```

```python
loaded_images = list()
for filename in listdir('Test/Images'):
    image = Image.open('Test/Images/' + filename)
    gs_image = image.convert(mode='L')
    img_resized = gs_image.resize((100,100))
    data = asarray(img_resized)
    data = data.reshape((10000,1))
    loaded_images.append(Pair(filename,data))
```

```python
df1=pd.read_csv("Test/test.csv")
for j in range(10000):
 df1["pixel"+str(j)]=   np.array([loaded_images[imag].second[j] for imag in range(len(loaded_images))])
#df1.drop('ImageId',axis=1)
```

```python
data1=df1.values
```

```python
X1=data1[:,2:]
```

```python
results=list()
for i in range(len(X1)):
 results.append(Pair(loaded_images[i].first,KNN(X,Y,X1[i])))
```
